```python
predictors = df1
target_names = target_names
print(predictors, '\n\nЦелевая переменная')
print(target)
print('Классы:\n', target_names)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(predictors, target, train_size = 0.8, shuffle = True, random_state = 271)
print('Размер для признаков обучающей выборки', x_train.shape, '\n',
     'Размер для признаков тестовой выборки', x_test.shape, '\n',
     'Размер для целевого показателя обучающей выборки', y_train.shape, '\n',
     'Размер показателя тестовой выборки', y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
import time
start_time = time.time()

model = LogisticRegression(random_state = 271)
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
print("Время работы алгоритма логистической регрессии: %s seconds" % (time.time() - start_time))
print('Предсказанные значения: ', y_predict)
print('Исходные значения: ', np.array(y_test))
```

```python
import plotly.express as px
from sklearn import metrics
from sklearn.metrics import confusion_matrix

fig = px.imshow(confusion_matrix(y_test, y_predict), text_auto = True)
fig.update_layout(xaxis_title = 'Target', yaxis_title = 'Prediction')
```

```python
from sklearn.metrics import classification_report
print('Результаты классификации Logistic Regression:\n\n', classification_report(y_test, y_predict))
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn import  model_selection
from sklearn.model_selection import GridSearchCV
start_time = time.time()

param_kernel = ('linear', 'rbf', 'poly', 'sigmoid')
parameters = {'kernel': param_kernel}
model = SVC()
grid_search_svm = GridSearchCV(estimator = model, param_grid = parameters, cv = 6)
grid_search_svm.fit(x_train, y_train)
best_model = grid_search_svm.best_estimator_
svm.preds = best_model.predict(x_test)
print("Время работы алгоритма SVM: %s seconds" % (time.time() - start_time))
print('Предсказанные значения: ', svm.preds)
print('Исходные значения: ', np.array(y_test))
```

```python
fig = px.imshow(confusion_matrix(y_test, svm.preds), text_auto = True)
fig.update_layout(xaxis_title = 'Target', yaxis_title = 'Prediction')
```

```python
print('Результаты классификации SVM:\n\n', classification_report(y_test, svm.preds))
```

```python
from sklearn.neighbors import KNeighborsClassifier
start_time = time.time()

number_of_neighbors = np.arange(3, 10, 20)
model_KNN = KNeighborsClassifier()
params = {"n_neighbors": number_of_neighbors}
grid_search = GridSearchCV(estimator = model_KNN, param_grid = params, cv = 6)
grid_search.fit(x_train, y_train)
knn_preds = grid_search.predict(x_test)
print("Время работы алгоритма KNN: %s seconds" % (time.time() - start_time))
print('Предсказанные значения: ', knn_preds)
print('Исходные значения: ', np.array(y_test))
```

```python
fig = px.imshow(confusion_matrix(y_test, knn_preds), text_auto = True)
fig.update_layout(xaxis_title = 'Target', yaxis_title = 'Prediction')
```

```python
print('Результаты классификации KNN:\n\n', classification_report(y_test, knn_preds))
```
